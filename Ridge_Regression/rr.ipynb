```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
data = pd.read_csv("../salary_data.csv")
```

```python

# Extract features and target
x = data['YearsExperience'].values.reshape(-1, 1)
y = data['Salary'].values.reshape(-1, 1)
```

```python
# Design matrix with bias term
X = np.hstack((np.ones((x.shape[0], 1)), x))

# Regularization parameter
lambda_ = 1000

# Identity matrix 
I = np.eye(X.shape[1])
I[0, 0] = 0
```

```python
# Ridge Regression Normal Equation
beta_ridge = np.linalg.inv(X.T @ X + lambda_ * I) @ X.T @ y
```

```python
# Parameters
beta_0 = beta_ridge[0, 0]
beta_1 = beta_ridge[1, 0]

print("Intercept (β0):", beta_0)
print("Slope (β1):", beta_1)
```

```python
# Predictions
y_pred = X @ beta_ridge

# Residuals
residuals = y - y_pred

# Ridge Loss Function
ridge_loss = (residuals.T @ residuals + lambda_ * (beta_ridge.T @ beta_ridge))[0, 0]
print("Ridge Loss:", ridge_loss)
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Ridge Regression (L2 Regularization)")
plt.show()
```
